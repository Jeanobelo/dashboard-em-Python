```python
from dash import Dash, html, dcc, Input, Output
import plotly.express as px
import pandas as pd
import webbrowser
import threading

USUARIOS = {"jeano": "123",
           "Jean": "321"}

# criar o aplicativo / dashboard
app = Dash(__name__)
# auth = dash_auth.BasicAuth(app, USUARIOS)

# assume you have a "long-form" data frame
# see https://plotly.com/python/px-arguments/ for more options
df = pd.read_excel(r"Vendas.xlsx")

fig = px.bar(df, x="Produto", y="Quantidade", color="ID Loja", barmode="group")
fig2 = px.scatter(df, x="Quantidade", y="Valor Final", color="Produto", size="Valor Unitário", size_max=60)

lista_marcas = list(df["Marca"].unique())
lista_marcas.append("Todas")
lista_paises = list(df["País"].unique())
lista_paises.append("Todos")

# layout
app.layout = html.Div(children=[
    html.H1(children='Meu Dashboard em Python'),

    html.Div(children='''
        Dashboard de Vendas em Python
    '''),

    html.H3(children='Vendas de cada Produto por Loja', id="subtitulo"),
    dcc.RadioItems(options=[{'label': marca, 'value': marca} for marca in lista_marcas], value="Todas", id="selecao_marcas"),
    dcc.Dropdown(options=[{'label': paises, 'value': paises} for paises in lista_paises], value="Todos", id="selecao_paises"),
    dcc.Graph(id='vendas_por_loja', figure=fig),
    dcc.Graph(id='vendas_por_loja2', figure=fig2),

], style={"text-align": "center"})

# Callbacks vai dar funcionalidade ao gráfico, vai filtrar os gráficos
@app.callback(
    Output("subtitulo", 'children'),
    Output("vendas_por_loja", 'figure'),
    Output("vendas_por_loja2", 'figure'),
    Input('selecao_marcas', 'value'),
    Input('selecao_paises', 'value'),
)
def selecionar_marca(marca, pais):
    if marca == "Todas" and pais == "Todos":
        texto = 'Vendas de cada Produto por Loja'
        fig = px.bar(df, x="Produto", y="Quantidade", color="ID Loja", barmode="group")
        fig2 = px.scatter(df, x="Quantidade", y="Valor Final", color="Produto", size="Valor Unitário", size_max=60)
    else:
        df_filtrado = df
        if marca != "Todas":
            # filtrar de acordo com a marca
            df_filtrado = df_filtrado[df_filtrado["Marca"] == marca]
        
        if pais != "Todos":
            # filtrar de acordo com o país
            df_filtrado = df_filtrado[df_filtrado["País"] == pais]
        
        texto = f'Vendas de cada Produto por Loja da {marca} e do {pais}'
        fig = px.bar(df_filtrado, x="Produto", y="Quantidade", color="ID Loja", barmode="group")
        fig2 = px.scatter(df_filtrado, x="Quantidade", y="Valor Final", color="Produto", size="Valor Unitário", size_max=60)
    
    return texto, fig, fig2

# Função para abrir o navegador
def open_browser():
    import time
    time.sleep(1)  # Espera um pouco para garantir que o servidor tenha iniciado
    webbrowser.open('http://127.0.0.1:8050/')

# colocar o seu site no ar
if __name__ == '__main__':
    # Inicia o servidor em um thread separado
    threading.Thread(target=lambda: app.run_server(debug=False)).start()
    open_browser()
```
